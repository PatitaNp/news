```python
from dsmlibrary.datanode import DataNode
import json
from tqdm.auto import tqdm
import pandas as pd
```

```python
dir_raw_id = 54
dir_process_id = 55
```

```python
datanode = DataNode()
```

```python
meta, fp = datanode.get_file(datanode.get_file_id(directory_id=dir_raw_id, name='news2.json'))
meta
```

```python
data = json.load(fp)
data
```

```python
spo = []
data = data
for news in tqdm(data):
        spo.append({
            'หมวดหมู่': news.get('topic', ""),
            'หัวข้อ': news.get('title', ""),
            'เนื้อหา' : news.get('detail', ""),
            'อ้างอิง' : news.get('url', ""),
            'สำนักข่าว' : news.get('news', "") 
        })
df = pd.DataFrame(spo)
df
```

```python
datanode.write(df=df, directory=dir_process_id, name="ข่าวเกี่ยวกับอาชีพ", profiling=True, lineage=[datanode.get_file_id(directory_id=dir_raw_id, name='news2.json')])
```
