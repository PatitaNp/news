```python
from bs4 import BeautifulSoup
import requests
import csv
import openpyxl
import json
import pandas as pd
import re
```

```python
topic_list = []
title_list = []
detail_list = []
date_list = []
project_href = []
```

```python
url = requests.get('https://www.ryt9.com/tag/%E0%B8%AD%E0%B8%B2%E0%B8%8A%E0%B8%B5%E0%B8%9E')
url.encoding = "uft-8"
soup = BeautifulSoup(url.text, 'html.parser') 
```

```python
for i in range(20):
    for c in soup.find_all('div', class_='news-list'):
        
        topic_list.append([item.text for item in c.find_all('span',{'class' : 'tag has-text-weight-bold is-gray is-medium'})][i])
        
        title_list.append([item.text for item in c.find_all('a',{'class' : 'list-title'})][i])

        detail_list.append([item.text for item in c.find_all('span',{'class' : 'brief'})][i])
        
        date_list.append([item.text for item in c.find_all('span',{'class' : 'date-time'})][i])
        
        project_href.append(c.a['href'])
            
print('Complete')
```

```python
table = pd.DataFrame([topic_list,title_list,detail_list,date_list,project_href]).transpose()
table.columns = ['topic','title','detail','date','url']
table.set_index('title')
```

```python
table.columns
table.dtypes
table.isna().sum().sort_values(ascending = False)
```

```python
table['news'] = table['date'].str.extract(r'([(\u0E00-\u0E7F)\s+.(\w+)]+$)')
```
